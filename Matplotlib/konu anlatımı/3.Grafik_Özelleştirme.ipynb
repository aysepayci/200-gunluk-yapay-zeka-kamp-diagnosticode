    Grafik Özelleştirme

```python
#Eksen başlıkları ve grafik başlığı ekleme
#Renk, işaretçi (marker), çizgi tipi (linestyle) ayarları
#Kılavuz çizgileri ekleme (grid)
#Eksen sınırlarını belirleme (xlim, ylim)
```

```python
import matplotlib.pyplot as plt 
import numpy as np
```

Eksen başlığı ekleme

```python
plt.xlabel("x label") #x ekseni için
plt.ylabel("y label") #y ekseni için
```

Grafik Başlığı ekleme

```python
x=[0,1,2,3,4,5,6]
y=[0,2,4,6,8,10,12]

plt.plot(x,y)
plt.axis([0,6,0,15])

plt.title("Grafik Başlığı")
plt.xlabel("x label")
plt.ylabel("y label")
plt.show()
```

ptl.plot parametreleri

```python
#'color':çizginin rengini belirler
#'linestyle' : çizgi stilini belirler
#'marker':noktalara denk gelen kesişimleri gösterir. "o" olabilir,"s" olabilir.
#'markersize':
# 'alpha':saydanlık-opaklık. 1=opak, 0=saydam
# 'label':grafiğin etiketi
```

```python
x=[0,1,2,3,4,5,6]
y=[0,2,4,6,8,10,12]

plt.plot(x,y,"o--r")
plt.axis([0,6,0,15])

plt.title("Grafik Başlığı")
plt.xlabel("x label")
plt.ylabel("y label")
plt.show()
```

```python
x=[0,1,2,3,4,5,6]
y=[0,2,4,6,8,10,12]

plt.figure(figsize=(6,4))
plt.plot(x,y,color="blue",marker="s",markersize=10,alpha=0.5)

plt.show()
```

```python
yil=[2011,2012,2013,2014,2015]

oyuncu1=[8,10,12,7,9]
oyuncu2=[7,12,5,15,21]
oyuncu3=[18,20,22,25,19]

plt.plot([],[],color="y",label="oyuncu1")
plt.plot([],[],color="r",label="oyuncu2")
plt.plot([],[],color="b",label="oyuncu3")
plt.stackplot(yil,oyuncu1,oyuncu2,oyuncu3,colors=["y","r","b"])
plt.title("yillara göre atılan goller")
plt.xlabel("yil")
plt.ylabel("gol sayisi")
plt.legend()
plt.show()
```

```python
#linestyle='--' kesik çizgili yaptık.
#marker='o' ile veri noktalarını belirginleştirdik.
#marker='x' ile veri noktalarını çarpı şeklinde yapabilirdik.
```

```python
gunler = [1, 2, 3, 4, 5]
notlar = [60, 75, 80, 85, 90]

# Çizgi grafiği oluşturma
plt.plot(gunler, notlar, color='b', linestyle='--', marker='o', markersize=8, label="Not Ortalaması")

plt.title("Öğrencinin Sınav Notları")
plt.xlabel("Gün")
plt.ylabel("Not")

plt.legend()

plt.show()
```

plt.scatter parametreleri

```python
#'color'=renk belirtir
#'label'=grafiğin etiketi
#'s'=noktaların büyüklüğü
#'edgecolors'=noktaları saran renk
#'alpha'=saydamlık ve opaklık
```

```python
x=np.arange(100)/10
plt.scatter(x,np.cos(x),color="red",label="Cosx",s=40,edgecolors="blue",linewidths=2,alpha=0.5)
plt.legend()
plt.show
```

plt.bar parametreleri

```python
#'color'= sütunların renk belirtir
# 'width'=sütunların genişliğini kontrol eder. 
#'edgecolors'=noktaları saran renk
#'linewidth'=kenarlık kalınlığını belirtir. 
```

```python
categories=["A","B","C","D","E"]
values=[35,20,45,51,10]

plt.bar(categories,values,color="red",width=0.5,edgecolor="blue",linewidth=2)
```

plt.hist() parametreleri

```python
#'bins':grup sayısını belirler
#'range':Bin'lerin alt ve üst liminitini belirler
#'density':True olursa grafik ypğunşuğunu belirtir,false olursa frkans belirtir.
#'color':Bin rengi
#'alpha':saydamlık opaklık
```

```python
data_normal=np.random.randn(10000)
plt.hist(data_normal,bins=30,color="magenta")
plt.show()
```

```python
plt.hist(data_normal,bins=30,color="magenta",range=(-3,3))
plt.show()
```

```python
#y ekseni değişti,yoğunluk gösteriyor,normalize değerler. 
plt.hist(data_normal,bins=30,color="magenta",range=(-3,3),density=True)
plt.show()
```

klavuz çizgisi ekleme

```python
x = [1, 2, 3, 4, 5]
y = [2, 4, 6, 8, 10]

plt.plot(x, y, marker='o', linestyle='-')

# Başlık ve etiketler ekleme
plt.title("Basit Çizgi Grafiği")
plt.xlabel("X Ekseni")
plt.ylabel("Y Ekseni")

# Kılavuz çizgileri ekleme
plt.grid(True)

plt.show()
```

```python
# xlim ve ylim ile aralık belirleyebiliriz. 
```

```python
x = [1, 2, 3, 4, 5]
y = [2, 4, 6, 8, 10]

plt.xlim(0,10)
plt.ylim(0,20)

plt.plot(x, y, marker='o', linestyle='-')

# Başlık ve etiketler ekleme
plt.title("Basit Çizgi Grafiği")
plt.xlabel("X Ekseni")
plt.ylabel("Y Ekseni")

# Kılavuz çizgileri ekleme
plt.grid(True)

plt.show()
```
