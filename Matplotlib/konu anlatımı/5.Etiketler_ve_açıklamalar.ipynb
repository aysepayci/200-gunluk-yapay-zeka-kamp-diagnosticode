    Etiketler ve açıklamalar

```python
#Legend (açıklama) ekleme
#Veri noktalarına etiket ekleme (annotate)
```

```python
import matplotlib.pyplot as plt 
import numpy as np
```

```python
x = [1, 2, 3, 4, 5]
y1 = [i**2 for i in x]  # y = x^2
y2 = [i**3 for i in x]  # y = x^3

plt.plot(x, y1, marker='o', linestyle='-', label="x^2", color="red")
plt.plot(x, y2, marker='s', linestyle='--', label="x^3", color="blue")

plt.xlabel("X Ekseni")
plt.ylabel("Y Ekseni")
plt.title("İki Fonksiyonun Grafiği")

plt.legend()  # Açıklama ekleyelim
plt.show()
```

```python
#label="x^2" ve label="x^3" kullanarak legend (açıklama) ekledik.
#plt.legend() fonksiyonuyla bu açıklamaları grafik üzerinde gösterdik.
```

```python
#plt.legend(loc="upper right") açıklama konumunun yerini ayarlama
```

```python
x = [1, 2, 3, 4, 5]
y = [i**2 for i in x]

plt.plot(x, y, marker='o', linestyle='-', label="y = x^2", color="green")

# Bir noktayı vurgulayalım
plt.annotate("Özel Nokta", xy=(3, 9), xytext=(3, 12),
             arrowprops=dict(facecolor='black', arrowstyle="->"))

plt.xlabel("X Ekseni")
plt.ylabel("Y Ekseni")
plt.title("Bir Noktayı Vurgulama")

plt.legend()
plt.show()
```

```python
#plt.annotate() fonksiyonuyla (3,9) noktasını işaretledik.
#arrowprops ile ok işareti ekledik.
```
