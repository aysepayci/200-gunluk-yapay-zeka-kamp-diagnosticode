```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x=np.linspace(-10,9,20)
y=x**3
z=x**2

fig,axes=plt.subplots(nrows=2,ncols=1,figsize=(4,4))

axes[0].plot(x,y)
axes[0].set_title("Cube")
axes[1].plot(x,z)
axes[1].set_title("Square")

plt.tight_layout()
fig.savefig("figure1.pdf")

plt.show()
```

```python
x = [1, 2, 3, 4, 5]
y = [i**2 for i in x]

plt.plot(x, y, marker='o', linestyle='-', color="purple")

plt.xlabel("X Ekseni")
plt.ylabel("Y Ekseni")
plt.title("Kaydedilen Grafik")

plt.savefig("grafik.png", dpi=300)  # Grafiği kaydet
plt.show()
```
