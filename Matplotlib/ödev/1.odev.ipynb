Grafik çizimi

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.arange(1, 6)
y1 = np.array([10, 15, 7, 12, 18])  
y2 = np.array([8, 13, 5, 10, 15])  

plt.figure(figsize=(8, 5))
```

```python
# Eksen başlıkları ve başlık
plt.xlabel("X Ekseni")
plt.ylabel("Y Ekseni")
plt.title("Çizgi ve Çubuk Grafiği")
```

```python
# Çizgi grafiği çizme
plt.plot(x, y1, marker='o', linestyle='--', color='green', label='Çizgi Grafiği')

# Açıklama (Legend) ekleme
plt.legend()

# Grafiği gösterme
plt.show()
```

```python
# Çubuk grafiği ekleme
plt.bar(x, y2, color='orange', alpha=0.6, label='Çubuk Grafiği')

# Açıklama (Legend) ekleme
plt.legend()

# Grafiği gösterme
plt.show()
```
