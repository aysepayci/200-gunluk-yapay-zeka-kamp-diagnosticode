```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Yeni figür oluştur (2 satır, 1 sütun)
fig, ax = plt.subplots(2, 1, figsize=(7, 8))
```

```python
# Rastgele veri oluştur
x = np.random.rand(50)
y = np.random.rand(50)
```

```python
# 1. Alt Grafik: Scatter Plot (Dağılım Grafiği)
axes[0].scatter(x, y, color='red', marker='o')
axes[0].set_title("Scatter Plot")
axes[0].set_xlabel("X Değerleri")
axes[0].set_ylabel("Y Değerleri")

# 2. Alt Grafik: Histogram
data = np.random.randn(1000)  # Normal dağılıma sahip rastgele veri
axes[1].hist(data, bins=30, color='purple', alpha=0.7)
axes[1].set_title("Histogram")
axes[1].set_xlabel("Değerler")
axes[1].set_ylabel("Frekans")

# Aralarındaki boşlukları ayarla
plt.tight_layout()

# Grafikleri göster
plt.show()
```
