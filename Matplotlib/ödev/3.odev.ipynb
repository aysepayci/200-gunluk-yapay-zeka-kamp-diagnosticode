```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.linspace(0, 10, 100)
y = np.sin(x)
```

```python
plt.figure(figsize=(7, 5))
plt.plot(x, y, color='blue', linewidth=8, linestyle="--")
plt.xlabel("X Ekseni")
plt.ylabel("Y Ekseni")
plt.title("Örnek Sinüs Grafiği")

# Grafiği kaydet (.png olarak)
plt.savefig("sinus_grafik.png", dpi=300)

# Grafiği göster
plt.show()
```
