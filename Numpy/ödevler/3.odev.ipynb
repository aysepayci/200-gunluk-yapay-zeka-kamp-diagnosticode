```python
import numpy as np
import matplotlib.pyplot as plt  #grafik çizmek için yazdık pc'nizde yoksa terminal'e pip install matplotlib yazarak indirebilirsiniz.
```

```python
sayi_arr=np.linspace(0,2*np.pi,100)
sayi_arr
```

```python
arr_sin=np.sin(sayi_arr)
print("dizi sinüs'ü =",arr_sin)
#alt alta yazmak için de bunu yazabilirsiniz
#for x in np.nditer(arr_sin):
#  print(x)

arr_cos=np.cos(sayi_arr)
print("dizi 'cosünü' =",arr_cos)
#for x in np.nditer(arr_cos):
#   print(x)
```

```python
# Grafiği çizdirme
plt.figure(figsize=(10, 6))
plt.plot(sayi_arr, arr_sin, label='sin(x)', color='blue')
plt.plot(sayi_arr, arr_cos, label='cos(x)', color='red')
```

```python
# Grafiğe başlık ve etiketler ekleme
plt.title('Sinüs ve Kosinüs Grafiği')
plt.xlabel('x (radyan)')
plt.ylabel('Değer')
plt.axhline(0, color='black', linewidth=0.5, linestyle='--')  # x ekseni çizgisi
plt.axvline(0, color='black', linewidth=0.5, linestyle='--')  # y ekseni çizgisi
plt.legend()
plt.grid(True, linestyle='--', alpha=0.7)

# Grafiği gösterme
plt.show()
```
