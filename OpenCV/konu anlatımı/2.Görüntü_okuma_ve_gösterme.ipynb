    Görüntü Okuma ve Gösterme

```python
#Resim dosyalarını okuma ( cv2.imread() )
#Görselleri gösterme ( cv2.imshow() )
#Görüntü boyutlarını ve özelliklerini öğrenme ( shape )
#Görseli farklı formatlarda kaydetme ( cv2.imwrite() )
```

```python
import  cv2
```

resim dosyası okuma ve gösterme

```python
kedi =cv2.imread('tatli_kedi.png') #imread parametre olarak dosya yolunu alır.
#kedi nesnesine atama yaptık ve sonra aşağıda kedi nesnesini gösterdik. 
```

```python
cv2.imshow("kedicik",kedi) #kedicik yerine farklı bir şey de yazabilirsiniz. 
cv2.waitKey(0) #0 olması beklemeden direkt başamasını sağlar. 
```

görüntü boyut bilgisi

```python
kedi.shape # (yükseklik,genişlik,kanal sayısı) görüntülenir
```

farklı kaydetme

```python
cv2.imwrite("tatli_kedi.jpg",kedi) #bu şekilde faklı kaydedilir
```
