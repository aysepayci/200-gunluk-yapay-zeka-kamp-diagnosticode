```python
import cv2
```

```python
kedicik=cv2.imread('tatli_kedi.png')
```

görüntüyü gri tonuna çevirme

```python
#cv2.cvtColor(resim, cv2.COLOR_BGR2GRAY)
```

```python
gri_kedicik=cv2.cvtColor(kedicik,cv2.COLOR_BGR2GRAY)
cv2.imshow("gri kedicik",gri_kedicik)
cv2.waitKey(0)
```

yeniden boyutlandırma

```python
#cv2.resize(resim, (genislik, yukseklik))
```

```python
cv2.imshow("kedicik",kedicik)
cv2.waitKey(0)
```

```python
kedicik.shape
```

```python
boyutlu_kedi=cv2.resize(kedicik,(350,250))
cv2.imshow("boyutlu kedi",boyutlu_kedi)
cv2.waitKey(0)
```

görüntüyü kırpma

```python
#kirpilan_resim = resim[baslangic_y:bitis_y, baslangic_x:bitis_x]
```

```python
kedicik.shape
```

```python
kirpik_kedicik=kedicik[50:150,50:150]
```

```python
cv2.imshow("kirpik kedi",kirpik_kedicik)
cv2.waitKey(0)
```

görüntü döndürme

```python
#cv2.rotate(resim, cv2.ROTATE_90_CLOCKWISE)
```

```python
dön_kedicik=cv2.rotate(kedicik,cv2.ROTATE_90_CLOCKWISE)
```

```python
cv2.imshow("dönük kedicik",dön_kedicik)
cv2.waitKey(0)
```

görüntü bulanıklaştırma

```python
#cv2.GaussianBlur(resim, (int, int), int)
```

```python
bulanik_kedi=cv2.GaussianBlur(kedicik,(9,5),0)
```

```python
cv2.imshow("bulanik kedi",bulanik_kedi)
cv2.waitKey(0)
```
