    Çizim ve Şekiller Ekleme

```python
#Çizgi çizme ( cv2.line() )
#Dikdörtgen ve çember ekleme ( cv2.rectangle() , cv2.circle() )
#Metin ekleme ( cv2.putText() )
```

```python
import cv2
import numpy as np
```

```python
siyah_sayfa=np.zeros((512,512,3),np.uint8) #siyah ortam oluşturma (boyut belirleme)
```

```python
cv2.imshow("siyah",siyah_sayfa)
cv2.waitKey(0)
```

```python
#(512,512,3) parametresi, görüntünün genişlik, yükseklik ve 3 kanallı (RGB) renk formatına sahip olduğunu belirtir.
#np.uint8 piksel değerlerinin 0 ile 255 arasında olmasını sağlar.
```

çizgi çizme

```python
#cv2.line(resim, (baslangic_x, baslangic_y), (bitis_x, bitis_y), (renk), kalinlik)
```

```python
cizgi=cv2.line(siyah_sayfa,(0,0),(siyah_sayfa.shape[1],siyah_sayfa.shape[0]),(0,255,0),3)
cv2.imshow("çizgi",cizgi)
cv2.waitKey(0)
```

dikdörgen çizme

```python
#cv2.rectangle(resim, (sol_ust_x, sol_ust_y), (sag_alt_x, sag_alt_y), (renk), kalinlik)
```

```python
dikdorgen=cv2.rectangle(siyah_sayfa,(0,0),(250,350),(0,0,255),cv2.FILLED) #dikdörtgen çekmek için fiiled doldurmak için kullunılır
cv2.imshow("dikdörtgen",dikdorgen)
cv2.waitKey(0)
```

çember ekleme

```python
#cv2.circle(resim, (merkez_x, merkez_y), yaricap, (renk), kalinlik)
```

```python
cember=cv2.circle(siyah_sayfa,(400,50),30,(255,255,0),5)
cv2.imshow("cember",cember)
cv2.waitKey(0)
```

metin ekleme

```python
#cv2.putText(resim, "Metin", (x, y), font, boyut, (renk), kalinlik)
```

```python
yazi=cv2.putText(siyah_sayfa,"kedi severim",(300,100),cv2.FONT_HERSHEY_COMPLEX,1,(110,50,50),1)
```

```python
cv2.imshow("yazi",yazi)
cv2.waitKey(0)
```
