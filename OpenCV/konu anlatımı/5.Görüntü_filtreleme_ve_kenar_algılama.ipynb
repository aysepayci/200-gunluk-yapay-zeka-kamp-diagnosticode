    Görüntü Filtreleme ve Kenar Algılama

```python
#Kenar belirleme ( cv2.Canny() )
#Eşikleme işlemi ( cv2.threshold() )
#Morfolojik işlemler (Erosion, Dilation)
```

```python
import cv2
import numpy as np
```

```python
kedi=cv2.imread('tatli_kedi.png')
```

kenar belirleme

```python
#Görüntü işleme, genellikle belirli nesneleri tespit etmek veya görüntüyü daha iyi anlamak için filtreleme ve kenar algılama tekniklerini kullanır.
#Canny kenar algılama, bir görüntüdeki keskin değişiklikleri (kenarları) belirlemek için kullanılan popüler bir tekniktir.
```

```python
#cv2.Canny(resim, esik1, esik2)
```

```python
kediCanny=cv2.Canny(kedi,150,200)       #gölgeleri silme, sadece siyah-beyaz çoğunlukla siyah
cv2.imshow("canny",kediCanny)
cv2.waitKey(0)
```

eşikleme

```python
#Eşikleme, bir görüntüyü siyah ve beyaz olarak bölmek için kullanılır.
```

```python
#cv2.threshold(resim, esik_degeri, max_deger, cv2.THRESH_BINARY)
```

```python
ret, thresh = cv2.threshold(kedi, 127, 255, cv2.THRESH_BINARY)
cv2.imshow('Esikleme', thresh)
cv2.waitKey(0)
```

Morfolojik İşlemler (Erosion, Dilation)

```python
#Morfolojik işlemler, nesnelerin şekillerini analiz etmek için kullanılır.
```

```python
#Erosion (Aşındırma) – Gürültüyü Azaltma
```

```python
kernel = np.ones((5,5), np.uint8)
eroded = cv2.erode(thresh, kernel, iterations=1)
cv2.imshow('Erosion', eroded)
cv2.waitKey(0)
```

```python
#Aşındırma, beyaz alanları küçültür, gürültüyü azaltır.
```

```python
#Dilation (Genişletme) – Beyaz Alanları Büyütme
```

```python
dilated = cv2.dilate(thresh, kernel, iterations=1)
cv2.imshow('Dilation', dilated)
cv2.waitKey(0)
```
