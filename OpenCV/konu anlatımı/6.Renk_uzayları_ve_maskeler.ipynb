    Renk Uzayları ve Maskeler

```python
#RGB, HSV ve diğer renk uzaylarına dönüşüm
#Belirli renkleri filtreleme (renk maskeleme)
#Görüntü segmentasyonu
```

```python
import cv2
import numpy as np
```

```python
kedi=cv2.imread('tatli_kedi.png')
```

```python
cv2.imshow('kedicik', kedi)
cv2.waitKey(0)
```

```python
#RGB’den HSV’ye Dönüştürme
 #HSV (Hue-Saturation-Value) renk uzayı, renkleri daha iyi ayırmamızı sağlar.
```

```python
hsv = cv2.cvtColor(kedi, cv2.COLOR_BGR2HSV)
cv2.imshow('HSV Görüntü', hsv)
cv2.waitKey(0)
```

```python
#Belirli Renkleri Filtreleme (Maskeleme)
 #Belirli bir rengi seçmek için cv2.inRange() fonksiyonunu kullanabiliriz.
```

```python
lower_red = np.array([0, 120, 70])
upper_red = np.array([10, 255, 255])

mask = cv2.inRange(hsv, lower_red, upper_red)
result = cv2.bitwise_and(kedi, kedi, mask=mask)

cv2.imshow('Kirmizi Maske', result)
cv2.waitKey(0)
```

```python
#lower_red ve upper_red, kırmızı rengin alt ve üst HSV sınırlarıdır.
```
