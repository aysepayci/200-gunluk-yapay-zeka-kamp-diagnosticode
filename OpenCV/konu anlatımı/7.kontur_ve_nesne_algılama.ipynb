```python
import cv2
```

```python
kedi=cv2.imread('tatli_kedi.png')
```

```python
cv2.imshow('kedi', kedi)
cv2.waitKey(0)
```

kontur bulma

```python
#cv2.findContours(resim, cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_SIMPLE)
```

```python
#cv2.findContours() renkli (3 kanallı) görüntüyle çalışmaz.
```

```python
# Görüntüyü oku ve gri tonlamaya çevir  
gri_kedi = cv2.cvtColor(kedi, cv2.COLOR_BGR2GRAY) 
```

```python
ret, thresh = cv2.threshold(gri_kedi, 127, 255, cv2.THRESH_BINARY)
cv2.imshow('tresh kedi', thresh)
cv2.waitKey(0)
```

```python
contours, _ = cv2.findContours(thresh, cv2.RETR_TREE, cv2.CHAIN_APPROX_SIMPLE)
```

kontur çizme

```python
#cv2.drawContours(resim, konturlar, -1, (renk), kalinlik)
```

```python
cv2.drawContours(kedi, contours, -1, (0, 255, 0), 2)
cv2.imshow('Konturlar', kedi)
cv2.waitKey(0)
```

Konturların Alanını ve Çevresini Hesaplama

```python
for contour in contours:
    area = cv2.contourArea(contour)
    perimeter = cv2.arcLength(contour, True)
    print(f"Alan: {area}, Çevre: {perimeter}")
```
