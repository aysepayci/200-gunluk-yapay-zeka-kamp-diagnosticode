```python
import cv2
import numpy as np
```

```python
insanlar=cv2.imread('insanlar.png')
```

```python
insanlar.shape
```

```python
insanlar=cv2.resize(insanlar,(512,512))
```

```python
cv2.imshow("insanlar",insanlar)
cv2.waitKey(0)
```

```python
#Konturlar ve Maskeleme:
#   Renk bazlÄ± maskeleme yap.
#   GÃ¶rseldeki konturlarÄ± tespit et ve Ã§iz.
```

```python
hsv_insanlar = cv2.cvtColor(insanlar, cv2.COLOR_BGR2HSV)
```

```python
cv2.imshow("hsv insanlar",hsv_insanlar)
cv2.waitKey(0)
```

```python
#zombilere benzediler biraz :)
```

```python
lower_red = np.array([0, 100, 100])   # KÄ±rmÄ±zÄ± iÃ§in alt sÄ±nÄ±r
upper_red = np.array([10, 255, 255])  # KÄ±rmÄ±zÄ± iÃ§in Ã¼st sÄ±nÄ±r
mask = cv2.inRange(hsv_insanlar, lower_red, upper_red)
```

```python
result = cv2.bitwise_and(insanlar, insanlar, mask=mask)
cv2.imshow("Maske result", result)
cv2.waitKey(0)
```

```python
lower_blue = np.array([90, 50, 50])    # Mavi iÃ§in alt sÄ±nÄ±r
upper_blue = np.array([130, 255, 255])  # Mavi iÃ§in Ã¼st sÄ±nÄ±r
mask = cv2.inRange(hsv_insanlar, lower_blue, upper_blue)
```

```python
result = cv2.bitwise_and(insanlar, insanlar, mask=mask)
cv2.imshow("maske resullt", result)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
lower_white = np.array([0, 0, 150])   # Beyaz iÃ§in alt sÄ±nÄ±r (dÃ¼ÅÃ¼k doygunluk ve yÃ¼ksek parlaklÄ±k)
upper_white = np.array([180, 50, 255])  # Beyaz iÃ§in Ã¼st sÄ±nÄ±r (yÃ¼ksek parlaklÄ±k)
mask = cv2.inRange(hsv_insanlar, lower_white, upper_white)
```

```python
result = cv2.bitwise_and(insanlar, insanlar, mask=mask)
cv2.imshow("Masked Image (White)", result)
cv2.waitKey(0)
```

```python
#YÃ¼z TanÄ±ma:
#   OpenCVânin hazÄ±r haarcascade_frontalface_default.xml modelini kullanarak yÃ¼zleri algÄ±la.
#   AlgÄ±lanan yÃ¼zleri dikdÃ¶rtgen iÃ§ine al.
```

```python
import cv2
faceCascade = cv2.CascadeClassifier(cv2.data.haarcascades + 'haarcascade_frontalface_default.xml')
```

```python
# Resmi yÃ¼kleyin
image = cv2.imread('insanlar.png')
cv2.imshow('Yuz Tanima', image)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
image.shape
```

```python
image=cv2.resize(image,(512,512))
```

```python
cv2.imshow('Yuz Tanima', image)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
cv2.imwrite('gray.png',gray)
cv2.imshow('Yuz Tanima', gray)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
#hatayÄ± bulamaadim ihÅallah bulurum :(
```

```python
# YÃ¼zleri tespit edin
faces = faceCascade.detectMultiScale(gray, scaleFactor=1.3, minNeighbors=3, minSize=(30, 30))

# YÃ¼zleri Ã§izin
for (x, y, w, h) in faces:
    cv2.rectangle(image, (x, y), (x+w, y+h), (255, 0, 0), 2)

# Sonucu gÃ¶sterin
cv2.imshow('Yuz Tanima', image)
cv2.waitKey(0)
cv2.destroyAllWindows()
```
