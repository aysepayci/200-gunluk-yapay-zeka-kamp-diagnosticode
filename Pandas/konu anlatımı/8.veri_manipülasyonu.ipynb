```python
#Amaç: 10 dolardan az ödediyse $, 10 ile 30 arasında ödediyse $$
# 30 dolar fazla ödediyse $$$ işareti bulunan fiyat katagorisi adında bir sütun oluşturun
```

```python
df.head()
```

```python
#1. yol
def fiyat_kategorisi(x):
    if x<10:
        return '$'
    elif 10<=x<30:
        return '$$'
    else:
        return '$$$'
```

```python
df['Fiyat_kategorisi']= df['total_bill'].apply(fiyat_kategorisi)
```

```python
df.sample(3)
```

```python
df.head()
```

```python
#2. yol
df['Fiyat_kategorisi']= df['total_bill'].apply(lambda x: '$' if x<10 else ('$$' if 10<=x<30 else '$$$'))
```

```python
df.sample(10)
```

```python
df.drop('Fiyat_kategorisi', axis=1, inplace=True)
```

```python
df['total_bill'].values
```

```python
df.head()
```

```python
def fiyat_kategori_vectorized(bill):
    categories=np.full_like(bill,'$',dtype='U3')

    categories[(bill>=10)&(bill<30)]='$$'

    categories[bill>=30]='$$$'

    return categories
```

```python
df['fiyat_kategorisi'] =fiyat_kategori_vectorized(df['total_bill'].values)
```
