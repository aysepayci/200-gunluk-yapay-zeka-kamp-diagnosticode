Gruplama ve Görselleştirme:

```python
import pandas as pd
import matplotlib.pyplot as plt 
```

```python
df = pd.read_csv('titanic.csv')
```

```python
# Yolcuları sınıf sütununa (Pclass) göre gruplama ve hayatta kalma oranlarını hesaplama
gruplu_hayatta_kalma = df.groupby("Pclass")["Survived"].mean() * 100
print("\nsinifa Göre Hayatta Kalma Oranlari:")
print(gruplu_hayatta_kalma)
```

```python

# Gruplara ait hayatta kalma oranlarını bir çubuk grafikle görselleştirme
plt.bar(gruplu_hayatta_kalma.index, gruplu_hayatta_kalma.values, color=["blue", "orange", "green"])
plt.title("Sinifa Göre Hayatta Kalma Oranları")
plt.xlabel("Pclass (Sinif)")
plt.ylabel("Hayatta Kalma Orani (%)")
plt.xticks(gruplu_hayatta_kalma.index)  
plt.show()
```
