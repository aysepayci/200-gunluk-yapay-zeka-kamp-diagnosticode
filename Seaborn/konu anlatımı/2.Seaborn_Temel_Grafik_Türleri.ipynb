```python
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns 
import numpy as np
```

```python
penguins = sns.load_dataset("penguins") #direkt internetten yüklüyoruz.
penguins
```

kategorik veri görselleştirme

```python
#Kategorik veriler, belirli gruplara ait verileri ifade eder. Örneğin, farklı şehirlerdeki ortalama sıcaklık veya çalışanların departmanlara göre maaşları kategorik verilere örnektir.
```

```python
# Çubuk Grafikleri (barplot)
#Kategorik değişkenler arasındaki farkı görmek için kullanılır.
# Varsayılan olarak ortalama değerleri gösterir.
```

```python
data = pd.DataFrame({
    'Kategori': ['A', 'B', 'C', 'D'],
    'Değer': [10, 20, 15, 25]
})

sns.barplot(x='Kategori', y='Değer', data=data)
plt.show()
```

```python
grouped_penquins=penguins.groupby(["species","sex"])["body_mass_g"].mean().reset_index()
grouped_penquins
```

```python
sns.barplot(data=grouped_penquins,
x="species",y="body_mass_g",hue="sex")
plt.show()
```

```python
# Kutu Grafikleri (boxplot)
#Veri dağılımını, ortanca değeri ve aykırı değerleri gösterir.
#Özellikle veri setindeki uç noktaları (outliers) görmek için kullanılır.
```

```python
sns.boxplot(x='Kategori', y='Değer', data=data)
plt.show()
```

```python
#Çizgi Grafikleri (lineplot)
#Zaman serileri gibi değişkenlerin zamanla nasıl değiştiğini gösterir.
```

```python
sns.lineplot(x=[1, 2, 3, 4, 5], y=[3, 5, 7, 6, 8])
plt.show()
```

2️⃣ Dağılım Grafikleri

```python
#Sürekli değişkenler arasındaki ilişkileri görmek için kullanılır.
```

```python
#Scatterplot (Saçılım Grafiği)
#İki değişken arasındaki ilişkiyi görmek için kullanılır.
```

```python
tips = sns.load_dataset("tips")
sns.scatterplot(x="total_bill", y="tip", data=tips)
plt.show()
```

```python
sns.scatterplot( 
    data=penguins,
    x="bill_length_mm",
    y="bill_depth_mm",
    hue="species")

plt.title("bill_length_mm vs bill_depth_mm")
plt.show()
```

```python
#Jointplot
#Hem dağılımı hem de scatterplot’u birlikte gösterir.
```

```python
sns.jointplot(x="total_bill", y="tip", data=tips, kind="scatter")
plt.show()
```

Yoğunluk ve Histogram Grafikleri

```python
#Bu grafik türleri, verinin dağılımını ve yoğunluğunu anlamak için kullanılır.
```

```python
#Distplot (Dağılım Grafiği)
#Bir değişkenin histogramını çizer.
```

```python
sns.histplot(tips["total_bill"], bins=20, kde=True)
plt.show()
```

```python
penguins = sns.load_dataset("penguins")
sns.histplot(data=penguins, x="flipper_length_mm",hue="sex")
```

```python
penguins = sns.load_dataset("penguins")
sns.histplot(data=penguins, x="flipper_length_mm", hue="species", multiple="stack")
```

```python
#KDE Plot (Kernel Density Estimate)
#Sürekli değişkenlerin olasılık yoğunluklarını gösterir.
```

```python
sns.kdeplot(tips["total_bill"], shade=True)
plt.show()
```
