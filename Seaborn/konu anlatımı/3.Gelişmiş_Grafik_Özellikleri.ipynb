```python
#cubehelix_palette()matplotlib varsayılanından biraz farklıdır; renk tonu tekerleğinin etrafında çok fazla dönmez veya çok geniş bir yoğunluk aralığını kapsamaz. 
```

```python
sns.cubehelix_palette(as_cmap=True)
```

```python
#Paletin nasıl görüneceğini kontrol etmek için diğer argümanlar cubehelix_palette().
#Değiştireceğiniz iki ana şey start(0 ile 3 arasında bir değer) ve rot, veya dönüş sayısıdır (rastgele bir değer, ancak genellikle -1 ile 1 arasında)
```

```python
sns.cubehelix_palette(start=.5, rot=-.5, as_cmap=True)
```

```python
sns.cubehelix_palette(start=2, rot=0, dark=0, light=.95, reverse=True, as_cmap=True)
```

```python
tips = sns.load_dataset("tips")

sns.set_palette("coolwarm")
sns.scatterplot(x="total_bill", y="tip", data=tips)
plt.show()
```

```python
 #Stil Seçenekleri
#Seaborn'un ön tanımlı stilleri şunlardır:

#whitegrid (Beyaz arka plan + grid çizgileri)
#darkgrid (Koyu grid çizgileri)
#white (Sade, grid çizgisiz)
#dark (Koyu arka plan)
```

```python
penguins = sns.load_dataset("penguins")
sns.scatterplot(
    data=penguins,
    x='flipper_length_mm',
    y='body_mass_g',
    hue='species'
)
plt.show()
```

```python
sns.set_style('darkgrid')
sns.scatterplot(
    data=penguins,
    x='flipper_length_mm',
    y='body_mass_g',
    hue='species',
    style='sex'
)
plt.title('Flipper Lengyh vs Body Mass')
plt.xlabel('Flipper Lenght (mm)')
plt.ylabel('Body Mass (g)')
plt.show()
```

```python
sns.set_style("whitegrid")
sns.boxplot(x="day", y="total_bill", data=tips)
plt.show()
```

Görsel Tema Ayarları


```python
#Tüm grafiklerin genel görünümünü değiştirebilmek için set_theme kullanabiliriz.
```

```python
sns.set_theme(style="darkgrid", palette="pastel")
sns.histplot(tips["total_bill"], kde=True)
plt.show()
```
