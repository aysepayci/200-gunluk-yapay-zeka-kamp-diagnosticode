```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import seaborn as sns 
```

Pairplot ile Çoklu Değişken Görselleştirme

```python
#Pairplot, veri setindeki bütün sayısal değişkenleri birbiriyle karşılaştıran dağılım grafikleri çizer.
#Değişkenler arasındaki ilişkileri incelemek için kullanılır.
#Hangi değişkenlerin güçlü bir korelasyona sahip olduğunu anlamaya yardımcı olur.
#Büyük veri setleri için hızlı bir genel bakış sağlar.
```

```python
# Seaborn'un hazır veri setlerinden biri olan iris'i yükleyelim
iris = sns.load_dataset("iris")

# Pairplot grafiğini çizelim
sns.pairplot(iris, hue="species", palette="husl")
plt.show()
```

```python
tips = sns.load_dataset("tips")

fig,axes=plt.subplots(2,2,figsize=(12,10))
sns.set_palette("coolwarm")
sns.scatterplot(x="total_bill", y="tip", data=tips,
ax=axes[0,0])

sns.set_style("whitegrid")
sns.boxplot(x="day", y="total_bill", data=tips,ax=axes[0,1])

sns.set_theme(style="darkgrid", palette="pastel")
sns.histplot(tips["total_bill"], kde=True,ax=axes[1,0])

plt.show()
```

Heatmap (Isı Haritası)

```python
#Değişkenler arasındaki ilişkinin gücünü gösterir.
```

```python
#Renkler, ilişkinin derecesini belirler.
#Koyu renkler → Güçlü ilişki
#Açık renkler → Zayıf ilişki
```

```python
np.random.seed(0)
uniform_data=np.random.rand(10,12)
```

```python
sns.heatmap(uniform_data)
```

```python
sns.heatmap(uniform_data,vmin=0,vmax=2)
```

```python
normal_data=np.random.randn(10,12)
sns.heatmap(normal_data,center=0)
```

```python
flights=sns.load_dataset("flights")
print(flights.columns)
```

```python
flights = flights.pivot(index="month", columns="year", values="passengers")
sns.heatmap(flights)
```

```python
#pivot() fonksiyonu, veriyi yeniden şekillendirmek için kullanılır.
#pivot() fonksiyonunun amacı, uzun (long) formattaki bir DataFrame’i geniş (wide) formata dönüştürmektir. 
#Özellikle, belirli bir sütunu satır indeksine, belirli bir sütunu sütun isimlerine ve belirli bir sütunu da değer olarak atamak için kullanılır.
```

```python
sns.heatmap(flights, annot=True, fmt="d", linewidths=.5, cmap="YlGnBu",cbar_kws={"orientation":"horizontal"})
```

```python
#annot=True → Hücrelerin içinde sayıların gösterilmesini sağlar.
#fmt="d" → Sayıları tam sayı (integer) formatında gösterir.
#linewidths=.5 → Hücreler arasındaki çizgilerin kalınlığını belirler.
#cmap="YlGnBu" → Isı haritasında sarı-yeşil-mavi renk paletini kullanır.
#cbar_kws={"orientation":"horizontal"} → renk barını aşağı yerleştik.
```
