```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
tips = sns.load_dataset("tips")
```

```python
df = pd.DataFrame({
    "Kategori": ["A", "B", "C", "D"],
    "Değer": [10, 25, 15, 30]
})

# Çubuk grafiği çiz
sns.barplot(x="Kategori", y="Değer", data=df, palette="viridis")
plt.show()
```

```python
#Bazen verileri kategorilere ayırarak analiz etmek isteriz. Seaborn’un hue, col ve row gibi parametreleri bu iş için kullanılır.
```

```python
sns.catplot(x="day", y="total_bill", hue="sex", data=tips, kind="bar")
plt.show()
```

```python
penguins = sns.load_dataset("penguins")
grouped_penguins=penguins.groupby(["species"])["body_mass_g"].mean().reset_index()
grouped_penguins
```

```python
sns.barplot(data=grouped_penguins,
                x="species",
                y="body_mass_g")
```
