```python
#Grafiklerin başlıklarını ve eksenlerini özelleştirmek için plt.xlabel(), plt.ylabel() ve plt.title() kullanılır.
```

```python
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns 
```

```python
tips = sns.load_dataset("tips")
```

```python
sns.scatterplot(x="total_bill", y="tip", data=tips)
plt.xlabel("Toplam Hesap ($)")
plt.ylabel("Bahşiş ($)")
plt.title("Toplam Hesap ve Bahşiş İlişkisi")
plt.show()
```

```python
#color="blue" ile rengi mavi yaptık.
#width=0.5 ile çubukların genişliğini azalttık.
```

```python
sns.barplot(x="day", y="total_bill", data=tips, color="blue", width=0.5)
plt.show()
```

```python
#Seaborn, grafiklerin boyutlarını ve yazı stillerini değiştirmek için context ayarlarını sunar.
```

```python
sns.set_context("poster")  # Daha büyük fontlar ve çizgiler kullan
sns.barplot(x="day", y="total_bill", data=tips)
plt.show()
```
