Kategorik Veri Görselleştirme

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#Bir Pandas DataFrame kullanarak bir barplot oluştur
```

```python
data = pd.DataFrame({
    'isimler': ["Ayşe", "isa", "elif", "efe", "zehra", "sena", "ahmet", "murat", "emre", "yunus",
                "ece", "nuri", "necip", "yakup", "mustafa", "Gülseren", "nisa", "sedat"],
    'cinsiyet': ['k', 'e', 'k', 'e', 'e', 'k', 'k', 'e', 'e', 'e', 'e', 'k', 'e', 'e', 'e', 'e', 'k', 'k'],
    'Notlar': [90, 100, 15, 25, 45, 10, 85, 75, 25, 5, 95, 100, 100, 55, 50, 80, 100, 75]
})

plt.figure(figsize=(12, 6))  
sns.barplot(data=data, x="isimler", y="Notlar", hue="cinsiyet")
plt.xticks(rotation=45)  # İsimlerin yatay sığması için döndür
plt.show()
```

```python
#Aynı veriyle bir boxplot çiz.
```

```python
plt.figure(figsize=(12, 6))  
sns.boxplot(data=data, x="isimler", y="Notlar")
plt.xticks(rotation=45)  # İsimlerin yatay sığması için döndür
plt.show()
```
