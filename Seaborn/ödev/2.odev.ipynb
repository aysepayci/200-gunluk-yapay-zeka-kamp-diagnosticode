Dağılım ve Yoğunluk Grafikleri

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#Bir dağılım grafiği çiz ve bu grafikte renk paleti kullan.
```

```python
data = pd.DataFrame({
    'isimler': ["Ayşe", "isa", "elif", "efe", "zehra", "sena", "ahmet", "murat", "emre", "yunus",
                "ece", "nuri", "necip", "yakup", "mustafa", "Gülseren", "nisa", "sedat"],
    'cinsiyet': ['k', 'e', 'k', 'e', 'e', 'k', 'k', 'e', 'e', 'e', 'e', 'k', 'e', 'e', 'e', 'e', 'k', 'k'],
    'Notlar': [90, 100, 15, 25, 45, 10, 85, 75, 25, 5, 95, 100, 100, 55, 50, 80, 100, 75]
})

plt.figure(figsize=(12, 6))  
sns.scatterplot(data=data, x="isimler", y="Notlar", hue="cinsiyet",palette="viridis")
plt.xticks(rotation=45)  
plt.show()
```

```python
data = pd.DataFrame({
    'isimler': ["Ayşe", "isa", "elif", "efe", "zehra", "sena", "ahmet", "murat", "emre", "yunus",
                "ece", "nuri", "necip", "yakup", "mustafa", "Gülseren", "nisa", "sedat"],
    'cinsiyet': ['k', 'e', 'k', 'e', 'e', 'k', 'k', 'e', 'e', 'e', 'e', 'k', 'e', 'e', 'e', 'e', 'k', 'k'],
    'Notlar': [90, 100, 15, 25, 45, 10, 85, 75, 25, 5, 95, 100, 100, 55, 50, 80, 100, 75]
})

plt.figure(figsize=(12, 6))  
sns.jointplot(data=data, x="isimler", y="Notlar", kind="scatter")
plt.xticks(rotation=45)  
plt.show()
```

```python
#Aynı veriyle bir distplot ve bir kdeplot çiz.
```

```python
sns.histplot(data=data, x="isimler",y="cinsiyet", hue="Notlar", multiple="stack")
plt.xticks(rotation=45)  
plt.show()
```

```python
plt.figure(figsize=(10, 5))  # Grafik boyutunu ayarla

sns.kdeplot(data=data, x="Notlar", hue="cinsiyet", fill=True, alpha=0.5, linewidth=2)

plt.title("Not Dağılımı KDE Grafiği", fontsize=14, fontweight='bold')
plt.xlabel("Notlar", fontsize=12)
plt.ylabel("Yoğunluk", fontsize=12)
plt.grid(True, linestyle="--", alpha=0.5)  # Arka plana grid ekle

plt.show()
```
