Heatmap

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

```python
#Bir DataFrame’den bir heatmap oluştur ve renk skalasını özelleştir
```

```python
# Rastgele sayılardan oluşan bir DataFrame oluşturma
data = np.random.rand(10, 10)  # 10x10 boyutunda bir matris oluştur
df = pd.DataFrame(data)
df
```

```python
# Isı haritasını oluşturma
plt.figure(figsize=(8, 6))  
sns.heatmap(df, cmap="viridis", annot=True, linewidths=.5)  # cmap ile renk paletini değiştir
plt.title("Rastgele Sayılardan Oluşan DataFrame Isı Haritası")
plt.show()
```

```python
# Isı haritasını oluşturma
plt.figure(figsize=(8, 6))  
sns.heatmap(df, cmap="plasma", annot=True, linewidths=.5)  # cmap ile renk paletini değiştir
plt.title("Rastgele Sayılardan Oluşan DataFrame Isı Haritası")
plt.show()
```

```python
# Isı haritasını oluşturma
plt.figure(figsize=(8, 6))  
sns.heatmap(df, cmap="cividis", annot=True, linewidths=.5)  # cmap ile renk paletini değiştir
plt.title("Rastgele Sayılardan Oluşan DataFrame Isı Haritası")
plt.show()
```
